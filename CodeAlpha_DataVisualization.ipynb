```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("test.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=20)
plt.title("Fare Distribution")
plt.show()
```

plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# Data Visualization Findings

1. Male passengers were more than female passengers.
2. Most passengers belonged to Class 3.
3. Majority of passengers were between 20 and 40 years old.
4. Fare values varied significantly among passengers.
5. Heatmap helped identify relationships among numerical features.
